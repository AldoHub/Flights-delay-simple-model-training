```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)

#read the data and all values that are "-" will be considered null values
flights = pd.read_csv('flights.csv', na_values=['-'])

#get the rows and columns
flights.shape

#shows the first set of 5 if no value is passed
#flights.head()
flights.head(10)

```

```python
#check for missing values
flights.isna().sum()
#inspect ARRIVAL_DELAY null values
flights[flights['ARRIVAL_DELAY'].isna()]
```

```python
#inspect relationship between null values
na_values = flights[flights['SECURITY_DELAY'].isna() & flights['AIRLINE_DELAY'].isna() & flights['LATE_AIRCRAFT_DELAY'].isna() & flights['WEATHER_DELAY'].isna()]
na_values.shape
```

```python
#histogram to visualize data better
na_values.hist(bins=60, figsize=(20,20))
```

```python
#create a filter, get all the records were ARRIVAL_DELAY is not missing
flights = flights[~flights['ARRIVAL_DELAY'].isna()]
flights.shape
```

```python
#check for any other missing info in the dataset
flights.isna().sum()
```

```python
#set a default value to the records/columns that are missing - in this case will be filled with a 0 value
flights[['AIR_SYSTEM_DELAY', 'SECURITY_DELAY', 'AIRLINE_DELAY', 'LATE_AIRCRAFT_DELAY', 'WEATHER_DELAY']] = flights[['AIR_SYSTEM_DELAY', 'SECURITY_DELAY', 'AIRLINE_DELAY', 'LATE_AIRCRAFT_DELAY', 'WEATHER_DELAY']].fillna(0)
flights
```

```python
#bins = granuality
flights.hist(bins=60,figsize=(20, 20))
```

```python
#create a new column with the sum of all delays
flights['all_delay'] = flights[['AIR_SYSTEM_DELAY', 'SECURITY_DELAY', 'AIRLINE_DELAY', 'LATE_AIRCRAFT_DELAY', 'WEATHER_DELAY']].sum(axis=1)
flights
```

```python
#if theres a delay return the name of the highest number column (probably cause of the delay in a flight), else return 'NO_DELAY'
flights['delay_reason'] = np.where(flights['all_delay'] > 0, flights[['AIR_SYSTEM_DELAY', 'SECURITY_DELAY', 'AIRLINE_DELAY', 'LATE_AIRCRAFT_DELAY', 'WEATHER_DELAY']].idxmax(axis=1), 'NO_DELAY' )
flights
```

```python
#see the values from the selected column
flights['delay_reason'].value_counts()
```

```python
#clean the data
flights = flights[['YEAR_', 'MONTH_', 'DAY_', 'DAY_OF_WEEK', 'AIRLINE', 'SCHEDULED_DEPARTURE', 'SCHEDULED_ARRIVAL', 'delay_reason']]
```

```python
#make the datasets
#X dataset
X_flights = flights[['YEAR_', 'MONTH_', 'DAY_', 'DAY_OF_WEEK', 'AIRLINE', 'SCHEDULED_DEPARTURE', 'SCHEDULED_ARRIVAL']]

#models can only read mathematical values, so make the AIRLINE columns with mathematical values representation
X_flights = pd.get_dummies(X_flights)

#y dataset
y_flights = flights[['delay_reason']]
```

```python
X_flights
```

```python
#data splitting for testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_flights, y_flights, test_size=0.20, random_state=15)
```

```python
#model training
from sklearn.tree import DecisionTreeClassifier
cfl = DecisionTreeClassifier().fit(X_train, y_train)
```

```python
#prediction using the test data set
y_pred = cfl.predict(X_test)
y_pred
```

```python
#score
cfl.score(X_test, y_test)
```

```python
#evaluation using other metrics
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred, labels=cfl.classes_)
cm
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize= (10, 7))
sns.heatmap(cm, annot=True, fmt='g', xticklabels=cfl.classes_, yticklabels=cfl.classes_)
```
